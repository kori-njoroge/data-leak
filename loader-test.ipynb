```python
import pandas as pd
import matplotlib.pyplot as plt

def detect_anomaly(data):
    # Set up the SPC chart
    count_column = 'linecount'  # Replace with the appropriate column name for your count data
    # Calculate the mean and standard deviation of the data
    mean = data[count_column].mean()
    std_dev = data[count_column].std()

    # Calculate the upper and lower control limits
    ucl = mean + 3 * std_dev
    lcl = mean - 3 * std_dev

    control_limits = (lcl, ucl)

    # Create the SPC chart
    fig, ax = plt.subplots()
    ax.plot(data[count_column], 'bo-', markersize=4)
    ax.axhline(y=mean, color='r', linestyle='--', linewidth=1, label='Mean')
    ax.axhline(y=ucl, color='g', linestyle='--', linewidth=1, label='UCL')
    ax.axhline(y=lcl, color='g', linestyle='--', linewidth=1, label='LCL')
    ax.set_xlabel('Sample Number')
    ax.set_ylabel(count_column)
    ax.legend()
    plt.show()

    # Detect anomalies
    anomalies = []
    for i, row in data.iterrows():
        if row[count_column] > control_limits[1] or row[count_column] < control_limits[0]:
            anomalies.append(row)

    # Send notification
    if len(anomalies) > 0:
        print("Anomalies detected:")
        for anomaly in anomalies:
            print(anomaly)
    else:
        print("No anomalies detected.")

# Example usage with your data
data = pd.DataFrame([
    {'linecount': 1},
    {'linecount': 2},
    {'linecount': 3},
    {'linecount': 4},
    {'linecount': 5},
    {'linecount': 6},
    {'linecount': 3},
    # Add more rows of data as needed
])
```

```python

detect_anomaly(data)
```
